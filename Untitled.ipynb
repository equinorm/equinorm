```python
l_dir = os.listdir('n_body_system/logs')
l_dir.sort()
```

```python
temp_l = ['exp-5-egnn_vel', 'exp-10-egnn_vel', 'exp-15-egnn_vel', 'exp-20-egnn_vel', 'exp-50-egnn_vel', 'exp-100-egnn_vel']
target_l = []
for l_dir_temp in l_dir:
    if temp_l[5] in l_dir_temp:
        target_l.append(os.path.join('n_body_system/logs', l_dir_temp, 'losess.json'))
```

```python
target_l = ['n_body_system/logs/exp-5-egnn_vel-1/losess.json',
 'n_body_system/logs/exp-10-egnn_vel-1/losess.json',
 'n_body_system/logs/exp-15-egnn_vel-1/losess.json',
 'n_body_system/logs/exp-20-egnn_vel-1/losess.json',
 'n_body_system/logs/exp-50-egnn_vel-1/losess.json',
 'n_body_system/logs/exp-100-egnn_vel-1/losess.json']
```

```python
data, name = [], []
for target_l_temp in target_l:
    with open(target_l_temp, "r", encoding="utf-8") as f:
        x = json.load(f)
        x_temp = np.array(x['losess'])
        x_temp[np.isnan(x_temp)] = np.inf
        data.append(x_temp)
        name.append(target_l_temp.split('/')[-2].split('-')[-1])
```

```python
plt.figure(figsize=(20,12))
plt.ylim(0, 0.05)
for data_temp in data:
    plt.plot(data_temp)
plt.legend(name)
plt.title('n_body=5')
plt.savefig('figure/n_body=5.pdf')
```
